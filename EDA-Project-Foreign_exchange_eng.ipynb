# 3) Let's start with the Euro column

We work on the column "EURO AREA - EURO / US $", change its name to "Euro". Divide the "Time Serie" column into days, months, years and days of the week to find out on which day of the week the data is most absent

````python
euro_dollar=exchange[["Time Serie","EURO AREA - EURO/US$"]]
euro_dollar.columns = ["Time Serie","Euro"]
euro_dollar.head()
````

````python
euro_dollar["Day"] = euro_dollar["Time Serie"].apply(lambda x: datetime.strptime(x,"%Y-%m-%d").day)
euro_dollar["Month"] = euro_dollar["Time Serie"].apply(lambda x: datetime.strptime(x,"%Y-%m-%d").month)
euro_dollar["Year"] = euro_dollar["Time Serie"].apply(lambda x: datetime.strptime(x, "%Y-%m-%d").year)
euro_dollar["Weekday"] = euro_dollar["Time Serie"].apply(lambda x: datetime.strftime(datetime.strptime(x, "%Y-%m-%d"), "%A"))
euro_dollar.drop("Time Serie", axis=1, inplace=True)
euro_dollar.head()
````

Let's calculate on which day of the week the data are most absent

````python
counter = []
for i in euro_dollar[euro_dollar["Euro"]=="ND"].index:
    counter.append(euro_dollar["Weekday"][i])
count=pd.Series(counter)
count.value_counts()
````

Dropping all rows with no data

````python
euro_dollar=(euro_dollar[euro_dollar['Euro']!="ND"])
euro_dollar.dropna()
euro_dollar['Euro']=euro_dollar['Euro'].astype(float)
euro_dollar
````

We check in which day of the week was the maximum and minimum value of Euro and the maximum and minimum values for each year

````python
max_euro_week_day = (euro_dollar[euro_dollar["Euro"]==euro_dollar["Euro"].max()]["Weekday"])
min_euro_week_day = (euro_dollar[euro_dollar["Euro"]==euro_dollar["Euro"].min()]["Weekday"])
print(f"Maximum value of Euro was in {max_euro_week_day[max_euro_week_day.index[0]]}")
print(f"Minimum value of Euro was in {min_euro_week_day[min_euro_week_day.index[0]]}")
````

````python
for i in euro_dollar["Year"].unique():
    print(f'{i}: min: {euro_dollar[euro_dollar["Year"]==i].min()["Euro"]}\tmax: {euro_dollar[euro_dollar["Year"]==i].max()["Euro"]}')
````

````python
max_euro_year = (euro_dollar[euro_dollar["Euro"]==euro_dollar["Euro"].max()]["Year"])
min_euro_year = (euro_dollar[euro_dollar["Euro"]==euro_dollar["Euro"].min()]["Year"])
print(f"Maximum value of Euro was in {max_euro_year[max_euro_year.index[0]]}")
print(f"Minimum value of Euro was in {min_euro_year[min_euro_year.index[0]]}")
````

Plotting a chart over the years for the Australian, New Zealand and Singapore dollars

````python
new_data=new_exchange[["Time Serie", "AUSTRALIAN DOLLAR", "NEW ZELAND DOLLAR", "SINGAPORE DOLLAR"]]
new_data
````

````python
new_data["Year"] = new_data["Time Serie"].apply(lambda x: datetime.strptime(x, "%Y-%m-%d").year)
new_data.drop("Time Serie", axis=1)
````

````python
new_data=new_data.astype({"AUSTRALIAN DOLLAR":"float", "NEW ZELAND DOLLAR":"float", "SINGAPORE DOLLAR":"float"})
````

````python
Australian = []
New_Zeland = []
Singapore = []
for i in new_data["Year"].unique():
    Australian.append(new_data[new_data["Year"]==i].mean()["AUSTRALIAN DOLLAR"])
    New_Zeland.append(new_data[new_data["Year"]==i].mean()["NEW ZELAND DOLLAR"])
    Singapore.append(new_data[new_data["Year"]==i].mean()["SINGAPORE DOLLAR"])
    
Australian = pd.DataFrame(Australian, columns=["Australian"])
New_Zeland = pd.DataFrame(New_Zeland, columns=["New_Zeland"])
Singapore = pd.DataFrame(Singapore, columns=["Singapore"])
Year = pd.DataFrame(new_data["Year"].unique(), columns=["Year"])
pd_data = pd.concat([Australian, New_Zeland, Singapore, Year], axis=1)
pd_data
````

````python
sns.set(style="ticks");
sns.set_context("talk");

for i in pd_data.columns[:3]:
    g = sns.catplot(x="Year", y=i, data=pd_data, kind='bar');
    g.set(xticks=[1,6,13,19], xticklabels=[2000,2006,2013,2019])
    g.ax.set_title(f"Information for {i} dollar")
````
